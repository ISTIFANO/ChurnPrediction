```python
from pyspark.sql import SparkSession
import pandas as pd
from pymongo import MongoClient
spark = SparkSession.builder.getOrCreate()

path =r"C:\Users\aamir\Desktop\YC\P\ChurnPrediction\src\infrastructure\data\clean_data.csv"
df_preprocessed = spark.read.csv(path, header=True, inferSchema=True)
df_preprocessed.show(5)
```

```python
df_pandas = df_preprocessed.toPandas()

records = df_pandas.to_dict("records")
```

```python
client = MongoClient("mongodb://localhost:27017/")
db = client["BankAttritionDB"]             
collection = db["clients"]   
```

```python
if len(records) > 0:
    collection.insert_many(records)
    print("Données prétraitées stockées avec succès dans MongoDB.")
else:
    print("Aucune donnée à insérer.")

client.close()
```
